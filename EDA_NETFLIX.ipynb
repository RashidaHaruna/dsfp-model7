## 1- Number of movies released in each decade from 1945 using a line chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('imdb_movies_shows.csv')
df.head(10)
```

```python
data = df.copy()
data
```

```python
df["release_year"].unique()
```

```python
# Add a new column 'Decade' to categorize movies into decades
df['Decade'] = (df['release_year'] // 10) * 10
```

```python
# Group the data by 'Decade' and count the number of movies in each decade
movies_per_decade = df.groupby('Decade').size()
print(movies_per_decade)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(movies_per_decade.index, movies_per_decade.values, marker='o', color='Lightgreen', linestyle='-')
plt.xlabel('Decade')
plt.ylabel('Number of Movies Released')
plt.title('Number of Movies Released per Decade (1945-2021)')
# Display the line chart
plt.show()
```

## 4- show bar chart for movie genres for children movies

```python
children_movies_list=['PG', 'G']
```

```python
# Filter the dataframe for G and PG rated movies
children_movies_df = df[df['age_certification'].isin(children_movies_list)].reset_index(drop=True)
```

```python
#children_movies_df['genres'][0].replace('[',"").replace("]"," ").replace('\'',"").strip().split(',')[0]
```

```python
def get_genre(x):
    x=str(x)
    x_genre = x.replace('[',"").replace("]"," ").replace('\'',"").strip().split(',')[0].strip()
    return x_genre
```

```python
children_movies_df['genre']=children_movies_df['genres'].apply(lambda x:get_genre(x))
```

```python
plt.figure(figsize=(12, 5))
light_green = "#90EE90"
regular_blue = "#3498db"
custom_palette = [light_green, regular_blue]
ax=sns.countplot(x=children_movies_df['genre'],hue=children_movies_df['age_certification'], palette=custom_palette)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title('Genre Count Distribution across Children Movies')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation = 90);
plt.show()
```

# 5 - Which of the genres have the highest runtime on the overall dataset and for each age group

```python
data.head()
```

```python
data['age_certification'].unique()
```

```python
import numpy as np
def categories(rating):
    if pd.isna(rating):
        return 'others'
    elif 'TV-14'in rating or 'PG-13' in rating or ' TV-Y7' in rating or 'NC-17' in rating:
        return 'Children'
    else:
        return 'Adults'
```

```python
data['Age groups'] = data['age_certification'].apply(categories)
data['Age groups']
```

```python
# if genres column is a string with multiple genres seperated by columns, split them into a list
#q2_df['genres'] = df['genres'].str.split(',')
q2_df = data

top_5_runtime_genres = (
    q2_df.explode('genres')
    .groupby(['Age groups', 'genres'])['runtime']
    .sum()
    .groupby('Age groups', group_keys=False)
    .nlargest(5)
    .reset_index()
)

top_5_runtime_genres
```
